```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# CSV 파일 읽기
df = pd.read_csv('/content/매출액.csv', index_col='date')

# 결과 저장용 데이터프레임
all_forecasts = []

# LSTM 모델을 위한 데이터 준비 함수
def prepare_data(series, n_steps):
    X, y = [], []
    for i in range(len(series)):
        end_ix = i + n_steps
        if end_ix > len(series) - 1:
            break
        seq_x, seq_y = series[i:end_ix], series[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)

# LSTM 모델 학습 및 예측 함수
def lstm_forecast(series, n_steps, n_forecast):
    scaler = MinMaxScaler(feature_range=(0, 1))
    series_scaled = scaler.fit_transform(series.values.reshape(-1, 1))
    X, y = prepare_data(series_scaled, n_steps)
    X = X.reshape((X.shape[0], X.shape[1], 1))

    model = Sequential()
    model.add(LSTM(50, activation='relu', input_shape=(n_steps, 1)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    model.fit(X, y, epochs=300, verbose=0)

    x_input = series_scaled[-n_steps:].reshape((1, n_steps, 1))
    forecast = []
    for _ in range(n_forecast):
        yhat = model.predict(x_input, verbose=0)
        forecast.append(yhat[0][0])
        x_input = np.append(x_input[:, 1:, :], yhat.reshape((1, 1, 1)), axis=1)

    forecast = scaler.inverse_transform(np.array(forecast).reshape(-1, 1)).flatten()
    return forecast

# 모든 열에 대해 LSTM 모델을 적용
n_steps = 3  # 과거 3년 데이터 사용
n_forecast = 5  # 향후 5년 예측
prediction_years = list(range(2023, 2028))

for column in df.columns:
    series = df[column]
    forecast = lstm_forecast(series, n_steps, n_forecast)

    forecast_df = pd.DataFrame({'date': prediction_years, column: forecast})
    all_forecasts.append(forecast_df.set_index('date'))

    # 예측 결과 시각화
    plt.figure(figsize=(10, 6))
    plt.plot(df.index, df[column], marker='o', label='실제 매출액')
    plt.plot(prediction_years, forecast, marker='x', linestyle='--', label='예측 매출액')
    plt.title(f'{column} 매출액 예측')
    plt.xlabel('연도')
    plt.ylabel('매출액')
    plt.legend()
    plt.grid(True)
    plt.show()

# 모든 예측 결과 병합
final_forecast = pd.concat(all_forecasts, axis=1)

# 성장률 계산
growth_rates = final_forecast.pct_change() * 100

# 성장률 및 예측 매출액 테이블 저장
result = pd.concat([df, final_forecast])
growth_rates.to_csv('lstm_growth_rates_and_forecasts.csv')

# 성장률 및 예측 매출액 테이블 출력
from IPython.display import display
display(result)

print("성장률 및 예측 매출액 테이블이 'lstm_growth_rates_and_forecasts.csv' 파일로 저장되었습니다.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 읽기
df = pd.read_csv('/content/매출액.csv', index_col=0)

# 인덱스를 datetime 형식으로 변환
df.index = pd.to_datetime(df.index, format='%Y')

# 업종별 매출액 시계열 그래프
plt.figure(figsize=(14, 7))
for column in df.columns:
    plt.plot(df.index, df[column], marker='o', label=column)
plt.title('업종별 매출액 시계열 그래프')
plt.xlabel('연도')
plt.ylabel('매출액')
plt.legend()
plt.grid(True)
plt.show()
```
